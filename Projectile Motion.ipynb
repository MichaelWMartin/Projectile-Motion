```python
import numpy as np
import math
import matplotlib.pyplot as plt
import pylab

# Import Libraries
```

```python
x=np.zeros(2000)
x[0]=0
y=np.zeros(2000)
y[0]=0
v=np.zeros(2000)
v[0]=78
g=9.8
C=0.5
mball=.04593
rho=1.225
r=.02133
A=3.14159*r**2
M=0.25
th=30
th=th*.0174
dt=.01
vx=np.zeros(2000)
vx[0]=v[0]*math.cos(th)
vy=np.zeros(2000)
vy[0]=v[0]*math.sin(th)
i=0

# Add initial conditions
```

```python
for i in range(2000):
    x[i+1]=x[i]+vx[i]*dt
    y[i+1]=y[i]+vy[i]*dt
    vx[i+1]=vx[i]-(C*rho*A/mball)*v[i]*vx[i]*dt-M*vy[i]*dt
    vy[i+1]=vy[i]-(C*rho*A/mball)*v[i]*vy[i]*dt+M*vx[i]*dt-g*dt
    v[i+1]=math.sqrt(vx[i]**2+vy[i]**2)
    if v[i]>14:
        C=7/v[i]
    if y[i+1]<0:
        break
        
# Loop for motion of golf ball
# x position
```

```python
plt.plot(x,y)
pylab.ylim([0,max(y)*1.05])
plt.title("Projectile Motion 2d")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.show()
print(max(x))
print(max(y))
print(v[i+1])
```
